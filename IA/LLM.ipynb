```python
import os
from docling.document_converter import DocumentConverter
```

```python
def load_file():
	all_text = '' #Armazenar o conteudo de todos os PDF's

	converter = DocumentConverter()

	for filename in os.listdir(os.getcwd()):
		if filename.endswith(".pdf"): #verifica o formato do arquivo
			path_pdf = os.path.join(os.getcwd(), filename)
			try:
				result = converter.convert(path_pdf)
				markdown_content = result.document.export_to_markdown()

				all_text += markdown_content + "\n\n"
			
			except Exception as e:
				print(f'Erro ao processar o arquivo {filename}: {e}')
```

```python
all_txt = load_file()
```

```python
import fitz
```

```python
def load_file(folder_path):
		all_text = ""  # Variável para armazenar o conteúdo de todos os PDFs		
		# Itera sobre todos os arquivos na pasta
		for filename in os.listdir(folder_path):
			if filename.endswith(".pdf"):  # Verifica se é um arquivo PDF
				path_pdf = os.path.join(folder_path, filename)
				pdf = fitz.open(path_pdf)

				# Lê o conteúdo do PDF
				full_txt = ''.join([pdf[page_num].get_text() for page_num in range(pdf.page_count)]).strip()
				all_text += full_txt + "\n\n"  # Adiciona o conteúdo do PDF ao texto total		
		return all_text
```

```python
import os
from langchain.chat_models import ChatOpenAI
from langchain.schema import SystemMessage, HumanMessage
from langchain.callbacks import get_openai_callback
```

```python
os.environ["OPENAI_API_KEY"] = ''
```

```python
def chat_gpt(prompt):
    #Inicializar o modelo
    llm = ChatOpenAI(temperature=0.5, model='gpt-4o-mini-2024-07-18')
    system_message = SystemMessage(content="""
        Você é um assistente de leitura de PDF altamente eficiente.
                                   
        Instruções:
        - Use o mesmo idioma em que a pergunta foi feita.
        - Tente priorizar o contexto fornecido.
                                   
        Sugestões:
        - Seja educado e se for apenas um agradecimento, só retribua.                                  	

	""")
    
	###
    user_message = HumanMessage(content=f'{prompt}')
	
    messages_to_send = [system_message, user_message]
    
    with get_openai_callback() as callback:
        response = llm(messages_to_send)
        custo = callback.total_cost
        
    return response.content, custo
```

```python
pergunta = 'Me faça um resumo do texto'
```

```python
prompt = f"Contexto: {all_txt} \n Pergunta: {pergunta}"
```

```python
resposta, custo = chat_gpt(prompt)
```
